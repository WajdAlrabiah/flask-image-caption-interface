```python
from flask import Flask, render_template, request
import requests

app = Flask(__name__)

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/classify_message', methods=['POST'])
def classify_message():
    user_message = request.form['user_message']

    """
    1. Logic for checking if the user_message is a link for an image
    or something else.

    change the following label to include the logic
    """
    label = "none" #or change to image_caption manually.

    if label == "image_caption":
        """
        2. Logic for the model inference for image_caption
        this should be called as an API
        """
        URL = "http://192.168.1.218:5001/generate_image_caption"
        # defining a params dict for the parameters to be sent to the API
        PARAMS = {'image_url': user_message}

        response_text = requests.post(url=URL, json=PARAMS)
        if response_text.status_code == 200:
            response_text = response_text.text
        else:
            response_text = "Error: Failed"
    else:
        response_text = "Hello, this app can only generate a caption"

    return render_template('index.html', response=response_text)

if __name__ == '__main__':
    app.run(host='127.0.0.1', port=5002)
```
